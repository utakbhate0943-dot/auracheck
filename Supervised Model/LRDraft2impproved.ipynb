```python
import sklearn as sk
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
import pandas as pd
import sklearn.feature_extraction
import sklearn.feature_selection
```

```python
data=pd.read_csv('C:\\Users\\takbh\\auracheck\\Dataset\\students_mental_health_survey_numeric.csv')
data.head()
data.info()
data.describe()
data.isnull().sum()
```

```python
target_cols = ['Depression_Score', 'Anxiety_Score', 'Stress_Level']

X = data.drop(columns=target_cols)
y = data[target_cols]

print('Input columns (X):')
print(X.columns.tolist())
print('\nDependent columns (y):')
print(y.columns.tolist())

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)

print('\nX_train shape:', X_train.shape)
print('X_test shape:', X_test.shape)
print('y_train shape:', y_train.shape)
print('y_test shape:', y_test.shape)
```

```python
# Depression score model (Logistic Regression) using pre-split data
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

predictor_cols = [
    'Diet_Quality',
    'Physical_Activity',
    'Semester_Credit_Load',
    'Substance_Use',
    'Extracurricular_Involvement'
]

target_col = 'Depression_Score'

# Use X_train/X_test and y_train/y_test from Cell 3, then filter selected columns
X_train_dep = X_train[predictor_cols].copy()
X_test_dep = X_test[predictor_cols].copy()
y_train_dep = y_train[target_col].astype(int)
y_test_dep = y_test[target_col].astype(int)

# Missing-value handling (fit on train, apply to test)
train_medians = X_train_dep.median(numeric_only=True)
X_train_dep = X_train_dep.fillna(train_medians)
X_test_dep = X_test_dep.fillna(train_medians)

dep_model = LogisticRegression(max_iter=2000)
dep_model.fit(X_train_dep, y_train_dep)

y_pred_train_dep = dep_model.predict(X_train_dep)
y_pred_test_dep = dep_model.predict(X_test_dep)

print('Predictor columns:', predictor_cols)
print('Dependent column:', target_col)

print('\n--- Training Data Performance ---')
print('Train Accuracy:', round(accuracy_score(y_train_dep, y_pred_train_dep), 4))
print('\nTrain Classification Report:\n')
print(classification_report(y_train_dep, y_pred_train_dep, zero_division=0))

print('\n--- Test Data Performance ---')
print('Test Accuracy:', round(accuracy_score(y_test_dep, y_pred_test_dep), 4))
print('\nTest Classification Report:\n')
print(classification_report(y_test_dep, y_pred_test_dep, zero_division=0))
```

```python
# Improved multi-target pipeline (all input columns + model selection)
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, f1_score, classification_report

# Ensure data/split exist
if 'data' not in globals():
    data = pd.read_csv(r'C:\Users\takbh\auracheck\Dataset\students_mental_health_survey_numeric.csv')

target_cols = ['Depression_Score', 'Anxiety_Score', 'Stress_Level']
if not all(c in data.columns for c in target_cols):
    raise ValueError('Required target columns not found in dataset.')

if not all(v in globals() for v in ['X_train', 'X_test', 'y_train', 'y_test']):
    X = data.drop(columns=target_cols)
    y = data[target_cols]
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42, stratify=y['Depression_Score']
    )

predictor_cols_all = X_train.columns.tolist()
X_train_all = X_train[predictor_cols_all].copy()
X_test_all = X_test[predictor_cols_all].copy()

print('Using all input columns:', len(predictor_cols_all))
print(predictor_cols_all)

results = []
best_models = {}

for target_col in target_cols:
    print('\n' + '=' * 90)
    print(f'Target: {target_col}')
    print('=' * 90)

    y_train_t = y_train[target_col].astype(int)
    y_test_t = y_test[target_col].astype(int)

    candidates = {
        'LogReg_balanced': (
            Pipeline([
                ('imputer', SimpleImputer(strategy='median')),
                ('scaler', StandardScaler()),
                ('clf', LogisticRegression(max_iter=5000, class_weight='balanced'))
            ]),
            {
                'clf__C': [0.3, 1.0, 3.0, 10.0]
            }
        ),
        'RandomForest_balanced': (
            Pipeline([
                ('imputer', SimpleImputer(strategy='median')),
                ('clf', RandomForestClassifier(random_state=42, n_jobs=-1, class_weight='balanced_subsample'))
            ]),
            {
                'clf__n_estimators': [300, 500],
                'clf__max_depth': [None, 12, 20],
                'clf__min_samples_leaf': [1, 2, 4]
            }
        )
    }

    best_name = None
    best_estimator = None
    best_test_f1 = -1

    for model_name, (pipe, param_grid) in candidates.items():
        search = GridSearchCV(
            estimator=pipe,
            param_grid=param_grid,
            scoring='f1_macro',
            cv=3,
            n_jobs=-1,
            verbose=0
        )
        search.fit(X_train_all, y_train_t)

        y_pred_train = search.predict(X_train_all)
        y_pred_test = search.predict(X_test_all)

        train_acc = accuracy_score(y_train_t, y_pred_train)
        test_acc = accuracy_score(y_test_t, y_pred_test)
        train_f1 = f1_score(y_train_t, y_pred_train, average='macro')
        test_f1 = f1_score(y_test_t, y_pred_test, average='macro')

        print(f"{model_name} | best_params={search.best_params_}")
        print(f"  Train Acc={train_acc:.4f}, Test Acc={test_acc:.4f}, Train F1_macro={train_f1:.4f}, Test F1_macro={test_f1:.4f}")

        results.append({
            'Target': target_col,
            'Model': model_name,
            'Train_Accuracy': train_acc,
            'Test_Accuracy': test_acc,
            'Train_F1_macro': train_f1,
            'Test_F1_macro': test_f1,
            'Best_Params': str(search.best_params_)
        })

        if test_f1 > best_test_f1:
            best_test_f1 = test_f1
            best_name = model_name
            best_estimator = search.best_estimator_

    best_models[target_col] = (best_name, best_estimator)

    y_best_pred = best_estimator.predict(X_test_all)
    print(f"\nBest model for {target_col}: {best_name}")
    print('Classification report (test):')
    print(classification_report(y_test_t, y_best_pred, zero_division=0))

results_df = pd.DataFrame(results).sort_values(['Target', 'Test_F1_macro'], ascending=[True, False])
print('\n' + '=' * 90)
print('Model comparison (sorted by Target, Test_F1_macro)')
print('=' * 90)
print(results_df[['Target', 'Model', 'Train_Accuracy', 'Test_Accuracy', 'Train_F1_macro', 'Test_F1_macro', 'Best_Params']])
```

```python
# Compact summary of improved model search results
summary_cols = ['Target', 'Model', 'Train_Accuracy', 'Test_Accuracy', 'Train_F1_macro', 'Test_F1_macro', 'Best_Params']

best_per_target = (
    results_df.sort_values(['Target', 'Test_F1_macro'], ascending=[True, False])
              .groupby('Target', as_index=False)
              .first()
)

print('Best model per target (by Test_F1_macro):')
print(best_per_target[summary_cols].to_string(index=False))

results_out = r'C:\Users\takbh\auracheck\Supervised Model\model_comparison_results.csv'
results_df.to_csv(results_out, index=False)
print(f'\nSaved full comparison table: {results_out}')
```
